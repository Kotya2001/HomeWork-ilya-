```python
import openpyxl

price_list = openpyxl.load_workbook('Downloads/pricelist (2).xlsx')
price_list_26 = openpyxl.load_workbook('Downloads/pricelist (3).xlsx')

sheet = price_list.active
sheet_26 = price_list_26.active

rows = sheet.max_row
rows26 = sheet_26.max_row

def division(sheet, sheet_26, rows, rows26):
    
    before_prices = [float(sheet.cell(row=i, column=3).value) for i in range(7, rows+1) if sheet.cell(row=i, column=3).value 
                    is not None]
    names_of_goods = [sheet.cell(row=j, column=2).value.strip() for j in range(7, rows+1) if sheet.cell(row=j, column=2).value 
                      is not None]
    today_prices = [float(sheet_26.cell(row=k, column=3).value) for k in range(7, rows26+1) if sheet_26.cell(row=k, column=3).value 
                    is not None]
    today_names_of_goods = []
    for price1, price2, good in zip(before_prices, today_prices, names_of_goods):
        if (price2 / price1) * 100 >= 5:
            today_names_of_goods.append(good)
            
    return today_names_of_goods
            
division(sheet, sheet_26, rows, rows26)  


```
